```python
#Автор: Дмитрий Петров

from numpy import *

#Загружаем тренировочные данные
q=open("spect.train","r")
train = [ a.strip().split(",") for a in q]
plus = [a for a in train if a[0]=="1"]
minus = [a for a in train if a[0]=="0"]
train_joined = [''.join(a) for a in train]

#print('len train:', len(train))
#print('len of one plus:', len(plus[0]), 'overall len plus:', len(plus))
#print('len of one minus:', len(plus[0]), 'overall len minus: ', len(minus))
q.close()

#Загружаем тестовые данные
q=open("spect.test","r")
test = [ a.strip().split(",") for a in q]
test_joined = [''.join(a) for a in test]
q.close

#print(len(test_joined))
#for i in test_joined:
#    print(i)

#test1 = ''.join(train[0])
#test2 = ''.join(train[1])
train_bin = []
#print(test1)
#print(bin(int(test1,2)))

for i in range(0,len(train)):
    temp = ''.join(train[i][1:])
    train_bin.append(int(temp,2))

#print(len(train_bin), len(train))
plus_bin = train_bin[0:39]
minus_bin = train_bin[40:79]

#print(type(plus_bin[1]))

#det_str = train[43][1:]
#print(det_str)
#to_detect = int(''.join(det_str))
#print(to_detect)

#Список голосов за каждый из тренировочных объектов
vote_list = []

#Функция фальсификации гипотез
def falsify(mask_bin, context_bin):
    for i in range(0,len(context_bin)-1):
        if mask_bin & context_bin[i] == mask_bin:
            return 0
        else:
            return 1

#Функция подсчета голосов для неизвестного объекта
def get_vote(to_detect, plus_bin, minus_bin):
    vote_count = 0
    for i in range(0,len(plus)-1):
        mask = to_detect & plus_bin[i]
        vote_count += falsify(mask, minus_bin)
    for i in range(0,len(plus)-1):
        mask = to_detect & minus_bin[i]
        vote_count -= falsify(mask, plus_bin)
    return vote_count

#Для каждого элемента из тренировочного набора считаем, сколько голосов за него
for i in range(0, len(train)):
    det_str = train[i][1:]
    to_detect = int(''.join(det_str))
    vote = get_vote(to_detect, plus_bin, minus_bin)
    vote_list.append(vote)
    #print('vote for object #', i, '('+''.join(det_str)+')', vote)
 
    
#ОЧЕНЬ ВАЖНЫЙ КУСОК
#Данные в этом наборе противоречивы: есть несколько одинаковых объектов, которые классифицируются в train и как "+", и как "-".
#Кроме того, в отрицательном контексте в принципе меньше признаков, поэтому он с большей вероятностью вложится в положительный.
#Поэтому я решил использовать вероятностный подходи и посмотрел распределение голосов для положительных и отрицательных примеров.
#Выяснилось, что для положительных оно имеет выборочное среднее 13.9487179487, и смещено вправо от него. 
#Для отрицательных оно имеет среднее 8.925 и смещено влево относительно него. 
#Поэтому в качестве решающего правила я взял количество голосов > 11. Почему именно 11? Потому что дисперсия положительных голосов
#оказалась на порядок меньше, чем у отрицательных, поэтому я решил, что нужно взять отсечение ближе к среднему положительных.

#print(mean(vote_list[0:39]),std(vote_list[0:39])/sqrt(40))
#print(mean(vote_list[40:89]),std(vote_list[40:89])/sqrt(40))

#Словарь, в котором мы храним количество правильно и неправильно угаданных данных. 
cv_res = {
 "11": 0,  #true positive
 "10": 0,  #false negative
 "01": 0,  #false positive
 "00": 0,  #true negative
 "NaN": 0, #unclassified
}

#Основная часть: смотрим на данные из теста и угадываем их.
for i in test_joined:
    det_str = i[1:]
    to_detect = int(''.join(det_str))
    vote = get_vote(to_detect, plus_bin, minus_bin)
    if vote > 11:
        decision = '1'
    else:
        decision = '0'
    if i[0] == '1' and decision == '1':
        cv_res['11'] +=1
    elif i[0] == '1' and decision == '0':
        cv_res['10'] +=1
    elif i[0] == '0' and decision == '1':
        cv_res['01'] +=1
    elif i[0] == '0' and decision == '0':
        cv_res['00'] +=1
    else:
        cv_res['NaN'] +=1

print('accuracy: ', (cv_res['11']+cv_res['00'])/(cv_res['11']+cv_res['00']+cv_res['01']+cv_res['10']+cv_res['NaN']))
print('true-positive: ', cv_res['11'], 'true-negative: ', cv_res['00'])
print('false-negative: ', cv_res['10'], 'false-positive: ', cv_res['01'])
print('uncertain:', cv_res['NaN'])

#Итоговая точность вышла 78,6%. Я считаю, что для ленивого классификатора это хороший результат, поскольку в описании данных сказано,
#что самый крутой (и существенно более сложный) алгоритм имеет точность около 84%. Это означает, что какую-то структуру мы уловили. 
#Кроме того, хорошо, что уровень false-positive составляет 3.6%. Для траспланталогии это важно, ведь пересадка 
#неподходящего органа вредит больше, чем отказ от потенциально плохой пересадки (в этом случае ищется подходящий донор).
```
